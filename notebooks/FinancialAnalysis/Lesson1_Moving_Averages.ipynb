# Python for Quantitative Analysis: Moving Averages




```python
import pandas as pd
import matplotlib.pyplot as plt
#import data from https://www.nasdaq.com/market-activity/quotes/historical
```

## Cleaning Up Data

```python
tsla = pd.read_csv('./data/TSLA.csv', sep=r'\s*,\s*', encoding='ascii', engine='python') #tesla stock
#replace currency with float values
tsla[tsla.columns[1:]] = tsla[tsla.columns[1:]].replace('[\$,]', '', regex=True).astype(float)
tsla = tsla.iloc[::-1] #reverse dataframe so dates are in ascending order
tsla['Date'] = tsla['Date'].convert_dtypes(convert_string=True)
tsla.index = tsla['Date'] #make date the index
print(tsla.tail)
```

## Calculate Moving Average

```python
tsla['PriceDiff'] = tsla["Close/Last"].shift(-1) - tsla["Close/Last"]
tsla['Return'] = tsla['PriceDiff'] / tsla['Close/Last']
tsla['Direction'] = [1 if tsla['PriceDiff'].loc[ei] > 0 else 0 for ei in tsla.index]

#50 day moving average
end_date = '06/11/2020'
start_date = '01/02/2019'

plt.figure(figsize=(10,8))

#plot moving average
MA_days_slow = 50
ma_label_slow = 'ma'+str(MA_days_slow)
tsla[ma_label_slow] = tsla['Close/Last'].rolling(MA_days_slow).mean()
tsla[ma_label_slow].loc[start_date:end_date].plot(label=ma_label_slow)

MA_days_fast = 10
ma_label_fast = 'ma'+str(MA_days_fast)
tsla[ma_label_fast] = tsla['Close/Last'].rolling(MA_days_fast).mean()
tsla[ma_label_fast].loc[start_date:end_date].plot(label=ma_label_fast)

#Plot closing prices
tsla['Close/Last'].loc[start_date:end_date].plot(label='Close')
plt.legend()
plt.show()
```

# Implementing a Trend-Following Trading Strategy

### Basic Ideas
  * **Fast signal**: moving average over a shorter timespan
  * **Slow signal**: moving average over a longer timespan
  * Hypotheis (TREND): if the fast signal exceeds the slow signal, the stock price will go up in the next several days. Otherwise the stock will decrease
  * Strategy: if the fast signal exceeds the slow signal, buy and hold one share of stock (long one share of stock)
  
### Experimental Setup
  * Alternative hypothesis 1 (HOLD): holding the stock for the entire duration would be just as if not more effective
  

```python
#Calculate shares to buy
tsla['Shares_TREND'] = [1 if tsla.loc[ei, ma_label_slow] < tsla.loc[ei,ma_label_fast] else 0 for ei in tsla.index ]
tsla['Profit_TREND'] = [tsla.loc[ei, 'PriceDiff'] * tsla.loc[ei,'Shares_TREND'] for ei in tsla.index]
tsla['Profit_TREND'].plot()

tsla['Shares_HOLD'] = [1 for ei in tsla.index ]
tsla['Profit_HOLD'] = [tsla.loc[ei, 'PriceDiff'] * tsla.loc[ei, 'Shares_HOLD'] for ei in tsla.index]
tsla['Profit_HOLD'].plot()

plt.axhline(y=0, color='red')
plt.show()
```

```python
# Calculate wealth
tsla['Wealth_TREND'] = tsla['Profit_TREND'].cumsum()
tsla['Wealth_HOLD'] = tsla['Profit_HOLD'].cumsum()

plt.figure(figsize=(20,16))

tsla['Wealth_TREND'].plot()
tsla['Wealth_HOLD'].plot()

plt.legend()
plt.show()

print(tsla.tail())
```
