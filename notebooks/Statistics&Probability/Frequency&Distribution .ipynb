# Frequency and Distribution

  * **frequency**: the number of occurences of an event in experiment or study
  * **distribution**: the description of the relative number of times each possible event will occur in a number of trials. Often describes the shape of the frequency bar graphs
  

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Set up model for example based on the dice roll game

die = pd.DataFrame([i for i in range(1,7)]) # create a 6-sided die
trial = 50
results = [die.sample(2, replace = True).sum().loc[0] for i in range(trial)] #simulate rolling 2 die 50 times

freq = pd.DataFrame(results)[0].value_counts()
sort_freq = freq.sort_index()
print(sort_freq)

sort_freq.plot(kind='bar', color = 'blue', figsize=(15,8))
```

# Relative Frequency

  * instead of counting the number of occurences, we are counting the percentage of occurences
  * do by dividing the number of occurences by number of trials

```python
#Now increase number of trials
die = pd.DataFrame([i for i in range(1,7)]) # create a 6-sided die
trial = 1000
results = [die.sample(2, replace = True).sum().loc[0] for i in range(trial)] #simulate rolling 2 die 50 times

freq = pd.DataFrame(results)[0].value_counts()
sort_freq = freq.sort_index()


# plot relative frequency, notice how it takes a more regular shape (like a normal distribution)
rel_freq = sort_freq/trial
rel_freq.plot(kind='bar',color='blue',figsize=(15,8))
```

# Describing Distribution

  * Mean and variance can be used to describe a distribution
 
### **variance** ($\sigma^2$): 
  * equal to the square of the standard deviation
  
$$ \sigma^2 = \frac{1}{N} \sum_{i=1}^{N}(x_i - \mu)^2 $$

  * N = number of trials
  * $x_i$ = random variable, the variable whose value is randomly determined for each trial
  * $\mu$ = mean
