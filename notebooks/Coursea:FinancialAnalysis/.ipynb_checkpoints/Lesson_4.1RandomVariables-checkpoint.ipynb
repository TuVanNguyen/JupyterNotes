# Association of Random Variables

## Covariance
  * measures the association between 2 variables

### Population Covariance
$
Cov(X,Y)=\frac{\sum_{i=1}^{N}(X_i - \mu_X)(Y_i - \mu_Y)}{N}
$

### Sample Covariance
$
Cov(X,Y)=\frac{\sum_{i=1}^{n}(X_i - \bar{X})(Y_i - \bar{Y})}{n-1}
$

## Coefficient of Correlation
  * covariance factorized to measure strength of association on scale of -1 to 1
  * both covariance and correlation can only address linear patterns
  * to analyze nonlinear patterns, we can use scatter plots

### Population Correlation Coefficient
$
\rho(X,Y) = \frac{Cov(X,Y)}{\sigma_X \sigma_Y}
$

### Sample Correlation Coefficient
$
r(X,Y) = \frac{Cov(X,Y)}{s_x s_y}
$

```python
import pandas as pd
import matplotlib.pyplot as plt

housing =  pd.read_csv('./data/housing.csv', sep=r'\s*,\s*', encoding='ascii', engine='python')
housing.head()
```

```python
#see correlation and covariance
print(housing.cov())
print(housing.corr())
```

```python
# now let's visually inspect relations between each pair of variables for other interesting patterns
from pandas.plotting import scatter_matrix
sm = scatter_matrix(housing, figsize=(10,10))
```

```python
# get a closer look at one of the relationships

housing.plot(kind='scatter', x='LSTAT', y='MEDV', figsize=(10,10))
```
